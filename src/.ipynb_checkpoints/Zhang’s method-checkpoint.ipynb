```python
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt 
```

## 1) Zhang Method For camara calibration 

```python


def zhangs():
    
    images = []
    
    for i in range(5456 ,5471):
        full_name = temp = '../calibration-data/IMG_'+str(i)+'.JPG'
        images.append(full_name)
    

    world_pts = []
    world_3d = [] 
    img_2d = [] 
    
    op_img = []
    
    for j in range(6):
        for i in range(8):
            world_pts.append([i,j,0])
    
    world_pts = np.asarray(world_pts,dtype=np.float32)

    for i in images:
        img = cv2.imread(i)
        img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        
        ret, corn = cv2.findChessboardCorners(gray, (8,6),None)

        subpix = cv2.cornerSubPix(gray,corn,(11,11),(-1,-1),(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001))
        
        op = cv2.drawChessboardCorners(img, (8,6), subpix, ret)
        op_img.append(op)
        
        world_3d.append(world_pts)
        img_2d.append(subpix)

    return world_3d,img_2d,gray,op_img

def plot_img(img):
    plt.figure(figsize=(20,20))

    for i in range(len(img)):
        plt.subplot(3, 5, i+1)
        plt.imshow(img[i])


images = get_img()
world_3d,img_2d, gray, op_img= zhangs()

plot_img(op_img)
```

```python
ret, mtx, dist,rvecs, tvecs = cv2.calibrateCamera( world_3d, img_2d, gray.shape[::-1],None,None)

ret
```

## 2) Wireframe Over Chess Board Points 

```python
img = cv2.imread('../calibration-data/IMG_5456.JPG')
img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)

dst,jacob = cv2.Rodrigues(rvecs[0])

p4 = np.column_stack((dst,tvecs[0]))
P = np.dot(mtx,p4)


proj_pts = []
for pts in world_3d[0]:
    temp = np.dot(np.array(P), np.append(pts ,1).T)
    temp = temp / temp[2]
    proj_pts.append((int(temp[0]),int(temp[1])))

img = cv2.circle(img, proj_pts[0], radius=1, color=(0,0,255), thickness=2)
for pts in range(len(proj_pts)-1):
    img = cv2.circle(img, proj_pts[pts+1], radius=1, color=(0,0,255), thickness=2) 
    img = cv2.line(img, proj_pts[pts], proj_pts[pts+1], color=(0,255,0), thickness=1) 
    
plt.figure(figsize=(15,10))
plt.grid(b=None)
plt.imshow(img)
plt.show();    
```

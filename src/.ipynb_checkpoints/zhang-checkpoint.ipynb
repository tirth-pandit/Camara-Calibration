#Zhangs Method


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def read_images(path):
  images = []
  for i in range(5456,5471):
    images.append(path + "IMG_"+str(i)+".JPG")
  return images

def zhangs(images):
  arr = []
  objpoints = [] 
  imgpoints = [] 
  show = []
  for i in range(6):
    for j in range(8):
      arr.append([j,i,0.0])
  arr = np.asarray(arr,dtype=np.float32)

  for i in images:
    img = cv2.imread(i)
    img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret, corners = cv2.findChessboardCorners(gray, (check_col,check_row),None)

    if ret == True:
        objpoints.append(arr)
        winsize = (11,11)
        zerozone = (-1,-1)
        corners2 = cv2.cornerSubPix(gray,corners,winsize,zerozone,(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001))
        img1 = cv2.drawChessboardCorners(img, (8,6), corners2, ret)
        show.append(img1)
        imgpoints.append(corners2)
  
  return objpoints,imgpoints,gray,show

def plotaxis(show):
  plt.figure(figsize=(20,15))

  for i in range(len(show)):
    plt.subplot(5, 3, i+1)
    plt.imshow(show[i])


path = '/content/drive/MyDrive/cv1/'
images = read_images(path)
objpoints, imgpoints, gray,show= zhangs(images)
```

# WireFrame over Project Points

```python
rotvec = np.column_stack(((cv2.Rodrigues(rvecs[0]))[0] ,tvecs[0]))
camera_matrix = np.dot(mtx,rotvec)
circle_color = (0,255,0)
line_color = (255,0,0)
thick = 6

def get_projection_points():
  x=[]
  proj= []
  for i in range(len(objpoints[0])):
    x.append(list(objpoints[0][i]))
    x[-1].append(1)

  x = np.asarray(x)
  for pts in x:
    temp = np.dot(np.array(camera_matrix), np.array(pts).T)
    proj.append([int(temp[0]/temp[2]),int(temp[1]/temp[2])])

  return proj

def project_rows(proj,img):
  for i in range(6):
    prevpts = tuple(proj[i*8])
    cv2.circle(img , tuple(prevpts), 1, circle_color,thick)
    for j in range(1,8):
      pts = tuple(proj[i*8+j])
      cv2.line(img,prevpts,pts,line_color,1)
      cv2.circle(img ,pts, 1, circle_color,4)
      prevpts = pts

def project_cols(proj,img):
  row_major = []
  for i in range(6):
    temp = []
    for j in range(8):
      temp.append(i*8+j)
    row_major.append(temp)

  for i in range(8):
    prevpts = tuple(proj[row_major[0][i]])
    for j in range(1,6):
      pts = tuple(proj[row_major[j][i]])
      cv2.line(img,prevpts,pts,line_color,1)
    
def wireframe():
  proj = get_projection_points()
  for i in range(6):
    for j in range(8):
      print(proj[i*8+j],end=" ")
    print(end="\n")
    
  img = cv2.imread('/content/drive/MyDrive/cv1/IMG_5456.JPG',cv2.IMREAD_COLOR)
  
  img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)
  prevpts = tuple(proj[0])
  cv2.circle(img,prevpts,1,circle_color,6)

  project_rows(proj,img)
  project_cols(proj,img)

  plt.imshow(img)
  plt.show()
```
